### Описание данных

В полных данных содержатся 19439 записей, содержащих в себе параметры добычи и очистки золота. По ТЗ:
- **Технологический процесс:**
    - Rougher feed — исходное сырье
    - Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
        - Xanthate — ксантогенат (промотер, или активатор флотации);
        - Sulphate — сульфат (на данном производстве сульфид натрия);
        - Depressant — депрессант (силикат натрия).
    - Rougher process (англ. «грубый процесс») — флотация
    - Rougher tails — отвальные хвосты
    - Float banks — флотационная установка
    - Cleaner process — очистка
    - Rougher Au — черновой концентрат золота
    - Final Au — финальный концентрат золота
- **Параметры этапов:**
    - air amount — объём воздуха
    - fluid levels — уровень жидкости
    - feed size — размер гранул сырья
    - feed rate — скорость подачи
- **Наименование признаков:**
Наименование признаков производится по такому шаблону:

[этап].[тип_параметра].[название_параметра]

Пример: `rougher.input.feed_ag`

Возможные значения для блока `[этап]`:
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

Возможные значения для блока `[тип_параметра]`:
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

### Расчёт эффективности

По ТЗ нам необходимо проверить, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака `rougher.output.recovery`. Эффективность обогащения рассчитывается по формуле:

<img src="https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg" align=left width=960, heigth=120>

  где:
- C — доля золота в концентрате после флотации;
- F — доля золота в сырье/концентрате до флотации;
- T — доля золота в отвальных хвостах после флотации.

Тогда, для нас:

```python
C = X_train.rougher_output_concentrate_au
F = X_train.rougher_input_feed_au 
T = X_train.rougher_output_tail_au

def eff(C, F, T):
    return 100 * (C * (F - T)) / (F * (C - T))
```

```python
display(X_train.rougher_output_recovery.isna().sum())
display(X_train.rougher_output_concentrate_ag.isna().sum())
display(X_train.rougher_input_feed_au.isna().sum())
display(X_train.rougher_output_tail_au.isna().sum())
```

Пропусков в проверяемых признаках нет, значит смело сравниваем по среднему абсолютному отклонению:

```python
print(f'MAE между рассчитанным и табличным значением эффективности обогащения на этапе флотации:\
 {mean_absolute_error(X_train.rougher_output_recovery, eff(C, F, T))}')
```

Что-то около нуля. Эффективность рассчитана верно.

### Создание пользовательской функции для вычисления sMAPE

Метрикой качества работы моделей по ТЗ определена sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»). Формула ее расчёта:

<img src="https://pictures.s3.yandex.net/resources/smape_1576238825.jpg" align=left width=960, heigth=120>


Так в библиотеке sklearn нет готовой функции для вычисления данной метрики, напишем её самостоятельно:

```python
def sMAPE(y_real, preds):
    '''
    Функция для расчёта симметричного среднего абсолютного процентного отклонение,
    между значениями векторов y_real и preds
    '''
    if len(y_real) != len(preds):
        raise Exception("Разное количество элементов в целевых и предсказанных значениях")
    return 100/len(y_real) * np.sum(2 * np.abs(y_real - preds) / (np.abs(y_real) + np.abs(preds)))
    
```

```python
print(sMAPE(np.array([1, 2]), np.array([1, 1])))
#sMAPE(np.array([1, 2]), np.array([3, 4, 5])) # в ексепшен вываливается, закоментил чтоб вас не пугать
```

```python
def score_dataset_cv(X_train, y_train, model, cv=5):
    """
    Тренирует модель на выборках X_train, y_train,
    с кросс-валидацией глубиной cv, возвращает sMAPE
    """
    scores = -1 * cross_val_score(model, X_train, y_train,
                             scoring=make_scorer(sMAPE, greater_is_better=False), cv=cv,n_jobs=-1)
    
    print(f'Результат sMAPE = {np.mean(scores):.4f}')
    return np.mean(scores)      
```

#### Подсчёт результатов моделей на тестовой выборке

Напишем функцию:

```python
def score_dataset(X_train, X_test, y_train, y_test, model):
    """
    Тренирует модель на выборках X_train, y_train,
    Предсказывает значения по X_test, возвращает sMAPE между получившимися знаениями и y_test.
    Для информации печатает MAE
    """
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    smape = sMAPE(y_test, preds)
    print(f'Результат sMAPE = {smape:.4f}, MAE = {mean_absolute_error(y_test, preds):.4f}')
    return smape     
```

```python
score_rougher = score_dataset(X_train_rougher, X_test_rougher, y_train_rougher, y_test_rougher, model_rougher)
```

```python
score_final = score_dataset(X_train_final, X_test_final, y_train_final, y_test_final, model_final)
```
