**Анализ продаж компьтерных игр.**

## Описание проекта

Из открытых источников доступны данные о продажах игр, оценках пользователей и экспертов, жанров и платформ. Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Цель исследования:**   
 Анализ данных о продажах игр за прошедшие периоды для выбора маркетинговой стратегии магазина компьтерных игр "Стричмик", для выявления потенциально популярных продуктов и планирования рекламных кампаний.

  
**Ход исследования:**

  Шаг 1. *О качестве данных ничего не известно. Поэтому перед исследованием понадобится обзор и предобработка полученных данных.*
   
   Шаг 2. *Для определения актуального периода, влияющего на продажи игр:*
   1) Определим, сколько игр выпускалось в разные годы. Выясним, важны ли данные за все периоды?
   2) Выясним, как менялись продажи по платформам. Определим платформы с наибольшими суммарными продажами и построим   распределение по годам. Обозначим  характерные сроки, за которые появляются новые и исчезают старые платформы.
   
   Шаг 3. *На основании выборки за актуальный для оценок продаж период проведем следующий анализ:*
   
   1) Обозначим, какие платформы лидируют по продажам, растут или падают. Выберем несколько потенциально прибыльных платформ.
   2) Построим диаграммы размаха по глобальным продажам игр в разбивке по платформам. Проанализируем результат.
   3) Проанализируем, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. Соотнесем с продажами игр на других платформах.
   4) Посмотрим на общее распределение игр по жанрам. Выявим самые прибыльные жанры. Выделим, при возможности, жанры с высокими и низкими продажами.
  
   Шаг 4. *Для пользователей каждого региона, определим:*
   
   1) Самые популярные платформы (топ-5). Опишем различия в долях продаж.
   2) Самые популярные жанры (топ-5). Опишем разницу для разных регионов.
   3) Определим влияет ли рейтинг ESRB на продажи в отдельном регионе.
    
   Шаг 5. *Проверим следующие гипотезы*
   1) Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
   2) Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
     
   Шаг 6. *По результатам проведенных исследований сформулируем общий вывод.*

### Колонка `rating`

```python
game_data.rating.value_counts(dropna=False)
```

Если почитать статью про рейтинг ESRB https://en.wikipedia.org/wiki/Entertainment_Software_Rating_Board

    E - everyone
    E10+ - everyone 10+
    T - teen
    M - mature
    AO - adults only
    
    EC - early childhood - старый рейтинг больще не используется.
    К-А - устаревшее название рейтинга E, поменяем его на E
    RP - rating pending, рейтинг проставляется, то есть его пока нет - тоже самое, что NaN.
    
Из этой же статьи узнаем, что данный рейтинг обязателен для продаж только на территории Северной Америки: Канады, США и Мексики и проставляться он начал только с 1994 года на добровольной основе. То есть пропуски связаны скорее всего с тем, что:

    1) Игра выпущена до 1994 года
    2) Производители игры не захотели оцениваться, в то время, когда рейтинг был добровольным, и его наличие не влияло на его размещение на площадках продаж.
    3) Рейтинг может отсутствовать на играх, продажи которых были ориентированы на не североамериканский рынок. Например, в Японии свой рейтинг - https://en.wikipedia.org/wiki/Computer_Entertainment_Rating_Organization, в Европе свой - https://en.wikipedia.org/wiki/PEGI и так далее. 
    
   Так что пропуски заменим на маркерное значение "without_rating"

Я поизучал справочную информацию в интернете по играм и по платформам и решил, что нужна еще одна категориальная колонка в данных: производитель платформы. Ведь нам нужно выяснить время жизни платформы, а значит надо знать когда обновляются поколения платформ, а поколения - это платформы одного производителя: 

     * MS: Xbox - Xbox360 - Xbox One
     * Sony: PS - PS2 - PS3 - PS4
       у них же портативная: PSP - PSV
     * Nintendo NES - SNES - N64 - GC - Wii -WiiU
       у них же портативная - GB - GBA - DS - 3DS
 То есть если мы построим графики продаж по годам, мы должны увидеть, как изменялись во времени продажи, как они перекрывались с друг-другом, когда производитель поддерживал обе своих платформы и увидим прекращения поддержки их. lets go.

```python
#напишем функцию категоризатор.
def developer_cat(platform:str) -> "str":
    microsoft_list = ['XB', 'X360', 'XOne']
    sony_main_list = ['PS', 'PS1', 'PS2', 'PS3', 'PS4']
    sony_portable_list = ['PSP', 'PSV']
    nintendo_main_list = ['NES', 'SNES', 'N64', 'GC', 'Wii', 'WiiU']
    nintendo_portable_list = ['GB', 'GBA', 'DS', '3DS']
    try:
        if platform in microsoft_list: return 'Microsoft'
        if platform in sony_main_list: return 'Sony_Main'
        if platform in sony_portable_list: return 'Sony_Portable'
        if platform in nintendo_main_list: return 'Nintendo_Main'
        if platform in nintendo_portable_list: return 'Nintendo_Portable'
        if platform == "PC": return 'PC' # отдельная не чисто игровая платформа
        return 'Other'
    except:
        print("Некорректные данные")
        return None    
```

```python
game_data['developer'] = game_data.platform.apply(developer_cat)
```

## Проверка гипотез

### Гипотеза о равенстве средних пользовательских рейтингов платформ Xbox One и PC.

Проверять будем игры из подборки игр за актуальный период из записей без пропусков в рейтинге.


Сформулируем нулевую и альтернативную гипотезу. Нулевая гипотеза H₀ должна предполагать равенство средних рейтингов, альтернативная гипотеза H₁ – это предположение, принимаемое в случае отклонения нулевой гипотезы:

H₀ - средние пользовательские рейтинги игр платформы XOne равны средним пользовательским рейтингам игр PC

H₁ - средние пользовательские рейтинги игр платформ XOne и PC различаются.

Критический уровень статистической значимости примем:

α = 5%

Для проверки гипотезы о равенстве средних оценок, воспользуемся методом `scipy.stats.ttest_ind (array1, array2, equal_var)` , где equal_var это параметр, указывающий равны ли дисперсии выборок. Для начала мы должны выяснить равны ли будут дисперсии наших выборок.

```python
user_score_xone = actual_games.query('user_score > 0 and platform == "XOne"').user_score
user_score_pc = actual_games.query('user_score > 0 and platform == "PC"').user_score
```

```python
print(f'Дисперсия выборки пользователских оценок игры платформы Xone = {np.var(user_score_xone)}')
print(f'Дисперсия выборки пользователских оценок игры платформы PC = {np.var(user_score_pc)}')
```

Дисперсии выборок значительно различаются. Поэтому мы примем equal_var=False

```python
alpha = .05
results = st.ttest_ind(
    user_score_xone, 
    user_score_pc, 
    equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Мы принимаем нашу нулевую гипотезу, а именно соглашаемся с тем, что средние пользовательские рейтинги игр платформы XOne равны средним пользовательским рейтингам игр PC. Попробуем визуализировать это.

```python
sns.catplot(data=actual_games.query('user_score > 0 and (platform == "XOne" or platform == "PC")'),
            x="platform", y='user_score', kind="box")
```

Да, значимых различий не наблюдается. Это может говорить о наличии кросс-платформенных игр, которые будут примерно одинаковы восприняты пользователями разных платформ
