# Финальный проект

## Описание проекта

В нашем распоряжении данные с параметрами обработки стали на металлургическом комбинате ООО «Так закаляем сталь». Необходимо оптимизировать производственные расходы путем уменьшения потребления электроэнергии на этапе обработки стали


**Цель проекта:**

Необходимо подготовить модель машинного обучения, предсказывающую темепратуру расплава стали на этапе достижения целевого химического состава. Целевой метрикой для нас определена средняя абсолютная ошибка MAE. По ТЗ её значение на тестовой выборке должно быть не больше 8.7

### План работ:

- 1. Подготовка данных
    - Откроем файлы и изучим их.
    - Проведем разведочный анализ (EDA) данных:
         - Проверим данные на пропуски, аномалии, выбросы, дубликаты, соответствия типов, наличие категориальных признаков и т.д.
         - Построим гистограммы распределений и диаграммы размаха.
         - Сформулируем вывод по EDA
         
         
- 2. Предобработка данных:
     - Обработаем пропуски и аномалии\выбросы. Обоснуем выбранные методы замены пропущенных\аномальных значений.
     - Определимся с признаками необходимыми для построения модели. При необходимости сформируем новые признаки из имеющихся данных.
     -  При обнаружении неинформативных признаков - удалим их.
     -  Объединим необходимые признаки в единый датасет из разных таблиц.
     -  Проверим данные на корреляцию. При необходимости сформируем для модели линейной регрессии датасет с учетом мультиколинеарности.
     -  Проведем масштабирование\нормализацию данных
     -  Сформулируем вывод по предобработке данных.
  
  
- 3. Построение моделей.
    - Разобьем данные на обучающую и тестовую выборку в пропорции 0.75\0.25.
    - Обучим разные модели и оценим их качество кросс-валидацией. Отберем лучшую. 
    - Проверим лучшую модель на адекватность dummy-регрессором.
    - Проверим метрику лучшей модели на тестовой выборке.
    
    
- 4. Сформулируем итоговый вывод

### Линейная регрессия.

Масштабирование сделаем внутри пайплайна, так будет правильнее:

```python
pipe = Pipeline([
        ('scale', MinMaxScaler()),
        ('reduce_dims', PCA(n_components=4)),
        ('lin_reg', LinearRegression(fit_intercept=True, normalize=False))])

scaler_list = [None, MinMaxScaler(), StandardScaler()]
param_grid = dict(reduce_dims__n_components=[x for x in range(22, 23)],
                  scale=scaler_list,
                  lin_reg__fit_intercept=[True, False],
                  lin_reg__normalize=[True, False])

grid = GridSearchCV(pipe, param_grid=param_grid, cv=5, n_jobs=1, scoring='neg_mean_absolute_error', verbose=5)
grid.fit(X_train, y_train)
score_lr = -grid.best_score_
model_final_lr = grid.best_estimator_
print(model_final_lr)
score_lr
```

Быстро и уже значение ниже ТЗ, посмотрим на модели посложнее.

```python
results = pd.DataFrame({'MAE_train': [np.round(score_lr, 4)]}, index=['logistic_regression'])
display(results)
```

### Случайный лес

```python
pipe = Pipeline([
    #('scale', MinMaxScaler()),
    #('reduce_dims', PCA(n_components=4)),
    ('forest', RandomForestRegressor(criterion='absolute_error', max_depth=21,
                      min_samples_leaf=4, min_samples_split=3, n_estimators=77,
                      random_state=RND_ST))
])

scaler_list = [None]
param_grid = {#'reduce_dims__n_components': [x for x in range(15, 27)],
              #'scale': scaler_list,
              'forest__n_estimators': [x for x in range(77, 78)],
              'forest__max_depth': [x for x in range(21, 22)],
              'forest__min_samples_split': [x for x in range(3, 4)],
              #'forest__max_features': [None, 1.0],
              'forest__criterion': ['absolute_error'],
              'forest__min_samples_leaf': [x for x in range(4, 5)],
              'forest__bootstrap': [True]}
grid = GridSearchCV(pipe, param_grid=param_grid, cv=5,
                    n_jobs=-1, scoring='neg_mean_absolute_error', verbose=5)
grid.fit(X_train, y_train)
score_rf = -grid.best_score_
model_final_rf = grid.best_estimator_
print(model_final_rf)
score_rf
```
